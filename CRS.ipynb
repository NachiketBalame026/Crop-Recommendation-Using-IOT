```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,:7],df['label'],test_size=0.2,random_state=42)
```

```python
import matplotlib.pyplot as plt
import asyncio
import websockets
import nest_asyncio
import re

nest_asyncio.apply()  # Apply the workaround for nested event loops

data_list = []

async def receive_data():
    uri = "ws://localhost:8765"  # Replace with actual IP or use localhost
    plt.ion()  # Enable interactive mode for live updating
    async with websockets.connect(uri) as websocket:
        while True:
            data = await websocket.recv()
            print(f"Received: {data}")

            # Use regex to extract numeric values
            match = re.search(r"Temp \(DHT\):\s*([0-9.]+)C", data)
            if match:
                temperature = float(match.group(1))  # Extract temperature and convert to float
                data_list.append(temperature)  # Store temperature

                # Optional: Remove old data if list size exceeds 50
                if len(data_list) > 50:
                    data_list.pop(0)

                plt.clf()  # Clear the current plot
                plt.plot(data_list)  # Plot the data
                plt.pause(0.1)  # Pause to update the plot with new data

# Run the receive_data function
await receive_data()  # Use `await` to call the function in Jupyter Notebook
```

```python
import asyncio
import websockets
import nest_asyncio
import re

nest_asyncio.apply()  # Apply the workaround for nested event loops

# Initialize lists to store the 10 readings
temperature_list = []
humidity_list = []
ph_list = []
soil_moisture_list = []  # Change rainfall to soil moisture

async def receive_data():
    uri = "ws://localhost:8765"  # Replace with actual IP or use localhost
    async with websockets.connect(uri) as websocket:
        while len(temperature_list) < 10:
            data = await websocket.recv()
            print(f"Received: {data}")

            # Extract temperature using regex
            temp_match = re.search(r"Temp \(DHT\):\s*([0-9.]+)C", data)
            if temp_match:
                temperature = float(temp_match.group(1))
                temperature_list.append(temperature)
            
            # Extract humidity using regex
            humidity_match = re.search(r"Humidity:\s*([0-9.]+)%", data)
            if humidity_match:
                humidity = float(humidity_match.group(1))
                humidity_list.append(humidity)
            
            # Extract pH using regex
            ph_match = re.search(r"pH:\s*([0-9.]+)", data)
            if ph_match:
                ph = float(ph_match.group(1))
                ph_list.append(ph)
            
            # Extract Soil Moisture (replacing rainfall)
            soil_moisture_match = re.search(r"Soil Moisture:\s*([0-9.]+)%", data)
            if soil_moisture_match:
                soil_moisture = float(soil_moisture_match.group(1))
                soil_moisture_list.append(soil_moisture)

            await asyncio.sleep(1)  # Adjust sleep if needed to control the frequency of readings

        # After collecting 10 readings, call the predict_crop function with the average values
        if len(temperature_list) == 10:
            avg_temperature = sum(temperature_list) / len(temperature_list)
            avg_humidity = sum(humidity_list) / len(humidity_list)
            avg_ph = sum(ph_list) / len(ph_list)
            avg_soil_moisture = sum(soil_moisture_list) / len(soil_moisture_list)  # Calculate average soil moisture

            # Call your predict_crop function with the average values
            predicted_crop = predict_crop(
                N=92, P=65, K=85,
                temperature=avg_temperature,
                humidity=avg_humidity,
                ph=avg_ph,
                rainfall=avg_soil_moisture  # Pass soil moisture as rainfall
            )

            # Print the prediction result
            print(f"Predicted Crop: {predicted_crop}")

# Run the receive_data function
await receive_data()  # Use `await` to call the function in Jupyter Notebook
```
